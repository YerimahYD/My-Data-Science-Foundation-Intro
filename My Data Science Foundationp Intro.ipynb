MACHINE LEARNING

The 3 main types of ML: SUPERVISED, UNSUPERVISED AND REINFORCED LEARNING

What supervised learning is

The difference between classification and regression

SUPERVISED LEARNING: TWO TYPES

A. Classification (LABELS[CATEGORY])

Output is a category

Examples:

Pass / Fail

Spam / Not spam

Yes / No


B. Regression (NUMBERS)

Output is a number

Examples:

House price

Temperature

Salary

FEATURES vs LABELS (FRAMING ML PROBLEMS CORRECTLY)

Clearly identify features

Clearly identify labels

Correctly frame a Machine Learning problem

If you get this wrong, models fail even with good code.

VERY IMPORTANT RULE

If the model already knows it before predicting → Feature

If the model is trying to predict it → Label

TRAINING DATA vs TEST DATA (AVOIDING CHEATING)

This is another critical concept. Many ML models fail because this is misunderstood.

Training data: Used to teach the model

Test data: Used to evaluate how well the model learned

The test data must be unseen during training

WHY WE SPLIT DATA

If you train and test on the same data:

Model looks perfect

But fails in real life

This is called overfitting.

| Term           | Meaning                |
| -------------- | ---------------------- |
| Training set   | Learning               |
| Test set       | Evaluation             |
| Overfitting    | Memorizing             |
| Generalization | Doing well on new data |


MODEL EVALUATION & ACCURACY

This is where Machine Learning stops being theory and starts being measurable.

Model evaluation measures how well predictions match real answers

Accuracy = correct predictions ÷ total predictions

95 emails are not spam

We must always understand:

The data

The problem

The consequences of mistakes

| Term            | Meaning               |
| --------------- | --------------------- |
| Accuracy        | Overall correctness   |

CONFUSION MATRIX (UNDERSTANDING MISTAKES)

A confusion matrix is just a scoreboard that counts:

How many times you were right

How many times you were wrong

How you were wrong


THE FOUR POSSIBLE OUTCOMES

For a Yes / No problem (like Spam / Not Spam):

| Name                | What it means                        |
| ------------------- | ------------------------------------ |
| True Positive (TP)  | You said **Yes**, and it **was Yes** |
| True Negative (TN)  | You said **No**, and it **was No**   |
| False Positive (FP) | You said **Yes**, but it **was No**  |
| False Negative (FN) | You said **No**, but it **was Yes**  |


PRECISION & RECALL (QUALITY OF PREDICTIONS)

PRECISION (HOW CAREFUL YOU ARE)

Precision answers:
When the model says YES, how often is it correct?

RECALL (HOW MUCH YOU CATCH)

Recall answers:
Out of all real YES cases, how many did we catch?

SIMPLE EXAMPLE (SPAM)

Model flags 10 emails as spam

Only 7 are actually spam

➡ Precision = low (many false positives)

There were 20 spam emails total

Model caught only 7

➡ Recall = low (many missed spam)

WHEN EACH ONE MATTERS (VERY IMPORTANT)

| Situation         | More Important |
| ----------------- | -------------- |
| Spam filter       | Precision      |
| Disease detection | Recall         |
| Fraud detection   | Recall         |
| Recommendations   | Precision      |


PRECISION vs RECALL TRADE-OFF (WHY MODELS CAN’T BE PERFECT)

This is where ML becomes realistic.

Imagine a guard checking students:

If the guard is very strict:

Few wrong people enter

But some real students are blocked

If the guard is very relaxed:

All students enter

But some strangers enter too

You must choose how strict to be.

That choice is the threshold.

THE TRADE-OFF (CORE IDEA)

Higher threshold:

 ↑ Precision

 ↓ Recall

Lower threshold:

 ↓ Precision

 ↑ Recall

You cannot maximize both at the same time.

Model outputs a cancer risk score:

0.0 → no risk

1.0 → high risk

If threshold = 0.9:

Very few false alarms

Many missed cases

If threshold = 0.3:

Catch almost all cancer

Many false alarms

WHY THIS MATTERS IN REAL LIFE

Medicine → prioritize recall

Spam → prioritize precision

Fraud → prioritize recall

ML decisions affect people, not just numbers.

UNDERFITTING vs OVERFITTING (BIAS & VARIANCE)

This is a core concept in Machine Learning.

Learning to Shoot a Ball 🎯

Underfitting: (Model is too simple, learn too little)

You barely practice → you miss a lot

Overfitting: (Model is too complex, memorize training data)

You memorize one exact spot → you miss everywhere else

Good fit: (Learns patterns that generalize)

You practice enough → you hit most targets

| Term           | Meaning                      |
| -------------- | ---------------------------- |
| Bias           | Error from being too simple  |
| Variance       | Error from being too complex |
| Generalization | Works on new data            |


| Situation                  | Problem       |
| -------------------------- | ------------- |
| Poor training + poor test  | Underfitting  |
| Great training + poor test | Overfitting   |
| Good training + good test  | Healthy model |

WHY THIS MATTERS

Overfitted models fail in real life

Underfitted models are useless

Balance is the goal of ML
